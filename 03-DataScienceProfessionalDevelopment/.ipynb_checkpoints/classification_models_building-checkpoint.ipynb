## 开始训练模型

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 选取特征列（X）和目标变量列（y）
X = df_cleaned.drop(columns=['PRICE_INTERVAL'])
y = df_cleaned['PRICE_INTERVAL']

# 将数据集分为训练集和测试集（70% 训练，30% 测试）
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 创建随机森林分类器模型
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# 在训练集上拟合模型
rf_model.fit(X_train, y_train)

# 在训练集上进行预测
train_predictions = rf_model.predict(X_train)

# 在测试集上进行预测
test_predictions = rf_model.predict(X_test)

# 计算在训练集和测试集上的准确度
train_accuracy = accuracy_score(y_train, train_predictions)
test_accuracy = accuracy_score(y_test, test_predictions)

# 输出训练集上的准确度
print(f"训练集准确度：{train_accuracy:.2f}")

# 输出测试集上的准确度
print(f"测试集准确度：{test_accuracy:.2f}")

# 输出其他性能指标（如分类报告和混淆矩阵）来评估模型性能
print("测试集性能报告：")
print(classification_report(y_test, test_predictions))

print("测试集混淆矩阵：")
print(confusion_matrix(y_test, test_predictions))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 创建随机森林分类器模型
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# 定义初始特征集，不包括目标变量 'PRICE_INTERVAL'
selected_features = list(df_cleaned.columns)
selected_features.remove('PRICE_INTERVAL')

# 划分数据集
X = df_cleaned[selected_features]
y = df_cleaned['PRICE_INTERVAL']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 计算原始特征集上的测试集准确度
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
best_accuracy = accuracy_score(y_test, y_pred)

# 开始逐步特征选择
while len(selected_features) > 1:
    worst_feature = None
    worst_accuracy = best_accuracy

    for feature in selected_features:
        # 暂时将特征从特征集中删除
        temp_selected_features = selected_features.copy()
        temp_selected_features.remove(feature)

        # 使用删除后的特征集训练模型
        X_temp = X[temp_selected_features]
        X_train_temp, X_test_temp, y_train_temp, y_test_temp = train_test_split(X_temp, y, test_size=0.3, random_state=42)

        rf_model.fit(X_train_temp, y_train_temp)
        y_pred_temp = rf_model.predict(X_test_temp)
        temp_accuracy = accuracy_score(y_test_temp, y_pred_temp)

        # 如果在测试集上的准确度提高，则更新最差特征和准确度
        if temp_accuracy > worst_accuracy:
            worst_feature = feature
            worst_accuracy = temp_accuracy

    # 如果找到特征可以提高准确度，则删除最差特征
    if worst_feature:
        selected_features.remove(worst_feature)
        best_accuracy = worst_accuracy
        print(f"删除特征 {worst_feature}，测试集准确度提高到 {best_accuracy:.2f}")
    else:
        break

# 输出最终保留的特征集
print("最终保留的特征集：", selected_features)
```
