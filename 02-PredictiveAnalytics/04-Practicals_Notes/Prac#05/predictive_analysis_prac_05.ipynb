# Evaluation Metrics Overview

there are four types of outcomes that could occur:
- True positives (TP) – when our model predicts that an observation belongs to a class and it actually belongs to that class.
- True negatives (TN) – when our model predicts that an observation does not belong to a class and it actually does not belong to a class.
- False positives (FP) – when our model predicts that an observation DOES belong to a class when it DOES NOT.
- False negatives (FN) – when our model predicts that an observation DOES NOT belong to a class when it DOES.

简单来讲：

- TP：模型预测某一个对象属于某一个class，并且这个对象确实属于这个class
- TN：模型预测某一个对象不属于某一个class，并且这个对象确实不属于这个class
- FP：模型预测某一个对象属于某一个class，但是这个对象并不属于这个class
- FN：某型预测某一个对象不属于某一个class，但是这个对象其实属于这个class

所有T开头的，都是正确的预测的，也就是True，而P和N则意味着是或者否，如果是则为P，如果不是则为N；反之则F。

## Building decision tree and calculating accuracy, precision, recall, and F1 score

````r
# load data
data <- read.csv(url("http://bit.ly/infs5100-stroke-data"))
````

````r
nrow(data)
ncol(data)
head(data)
````

````r
# set variable types
data$stroke <- as.factor(data$stroke)
data$gender <- as.factor(data$gender)
data$hypertension <- as.factor(data$hypertension)
data$heart_disease <- as.factor(data$heart_disease)
data$ever_married <- as.factor(data$ever_married)
data$work_type <- as.factor(data$work_type)
data$residence_type <- as.factor(data$residence_type)
data$smoking_status <- as.factor(data$smoking_status)
data$bmi <- as.numeric(data$bmi)
````

````r
library("mice")
# Impute missing BMI values
data.imputed <- mice(data, m=3, maxit = 50, method = 'pmm', seed = 500, printFlag = FALSE)
data.complete <- complete(data.imputed, 1)
````

````r
# Create training and test sets
set.seed(1000)
data.class <- data.complete[, c(-1)]  # c(-1)为不包含第一列，也就是id

train_index <- sample(1:nrow(data.class), 0.8 * nrow(data.class))
test_index <- setdiff(1:nrow(data.class), train_index)
train <- data.class[train_index,]
test <- data.class[test_index,]
list( train = summary(train), test = summary(test) )
````

````r
# Build full decision tree
library(rpart)
c.tree.full <- rpart(stroke ~ ., train, method = "class", cp=0)
````

````r
# Prune decision tree
p.tree.prune <- prune(c.tree.full, cp= c.tree.full$cptable[which.min(c.tree.full$cptable[,"xerror"]),"CP"])
````

````r
# Make a prediction
stroke.predict <- predict(p.tree.prune, test, type = "class")
````

````r
# Print confusion matrix
table(stroke.predict, test$stroke)
````

````r
# accuracy
accuracy <- sum(stroke.predict == test$stroke) / length(test$stroke)
accuracy
````

````r
TP <- sum(stroke.predict == 1 & test$stroke == 1)
FP <- sum(stroke.predict == 1 & test$stroke == 0)
precision <- TP / (TP + FP)
precision
````

````r
FN <- sum(stroke.predict == 0 & test$stroke == 1)
recall <- TP / (TP + FN)
recall
````

````r
f1_score <- 2 * (precision * recall) / (precision + recall)
f1_score
````

# The ROC Curve

````r
# load package
library(pROC)

# calculating and plotting ROC
prob.stroke = predict(p.tree.prune, newdata = test, type = "prob")[,2]

res.roc <- roc(test$stroke, prob.stroke)
plot.roc(res.roc, print.auc = TRUE)
````

在这段代码中，假设你已经训练了一个决策树模型并使用该模型对测试数据集进行了预测。p.tree.prune是决策树模型的对象，test是测试数据集。predict函数用于生成预测概率，type = "prob"表示返回概率而不是分类结果，[,2]表示提取预测结果中的第二列，即阳性类别的概率。

然后，使用roc函数来计算ROC曲线。roc函数的第一个参数是真实的分类标签，这里假设在test数据集中，阳性类别的标签存储在test$stroke中。第二个参数是预测的概率，即prob.stroke。

最后，使用plot.roc函数绘制ROC曲线，并通过设置print.auc = TRUE来打印AUC值（曲线下面积）。

请确保在运行代码之前已经安装并加载了pROC包。

请注意，在这里我们使用`type = "prob"`来获取一个类概率的矩阵，而不是之前使用的预测值。
我们可以注意到，ROC曲线表示敏感性和特异性之间的比率。灰色的对角线代表一个不比随机猜测更好的分类器。一个性能良好的分类器会有一个急剧上升到左上角的ROC曲线，即它会正确识别许多阳性样本，同时尽量不将阴性样本误分类为阳性。

在我们的例子中，AUC为0.784，远远好于随机猜测，表明我们建立了一个非常好的分类器。

我们还可以通过寻找使敏感性和特异性之和最大的最佳阈值来进一步优化分类器。请注意，一般情况下可能会有多个具有最大敏感性和特异性之和的最佳阈值。

# Cross Validation

交叉验证是一种重采样方法，它使我们能够获得可概括且更真实的错误率估计。 在 R 中运行交叉验证的方法有很多。我们将使用一种相当简单的方法，即利用 caret 包来构建与上面相同的决策树。

````r
library(caret)
````

````r
library(rpart.plot)
````

````r
# Set up caret to perform 10-fold cross validation
cv.control <- trainControl(method = "cv", number = 10)
````

trainControl函数用于创建一个交叉验证控制对象，以指定交叉验证的设置。具体解释如下：

method = "cv"：表示使用交叉验证方法进行模型评估。这里选择了常见的K折交叉验证（K-fold cross-validation）方法，其中K的值由number参数指定。

number = 10：表示将数据集划分为10个子集，每次使用其中9个子集作为训练集，剩下的一个子集作为验证集进行模型评估。这样的过程将重复进行10次，每次使用不同的子集作为验证集。最终，模型的性能评估结果将基于这10次交叉验证的平均值。
通过使用交叉验证，可以更准确地评估模型的性能，避免对单一训练集的过拟合，并对模型的泛化能力进行评估。

接着build a decision tree：

````r
# Use caret to train the rpart decision tree using 10-fold cross
# validation and use 15 values for tuning the cp parameter for rpart.
# This code returns the best model.
rpart.cv <- train(stroke ~ ., data = train, method = "rpart", trControl = cv.control, tuneLength = 15)
````

这段代码使用`caret`包中的`train`函数来训练`rpart`决策树模型，并结合10折交叉验证和调优参数`cp`的15个取值，最终返回最佳模型。

下面是对代码的解释：

```R
# 使用10折交叉验证和15个取值来训练rpart决策树模型
rpart.cv <- train(stroke ~ ., data = train, method = "rpart", trControl = cv.control, tuneLength = 15)
```

在这段代码中，`train`函数用于训练机器学习模型。具体解释如下：

- `stroke ~ .`：表示将`stroke`作为响应变量，其余的变量（除去响应变量）作为预测变量。
- `data = train`：表示使用`train`数据集进行训练。
- `method = "rpart"`：表示选择`rpart`方法来构建决策树模型。
- `trControl = cv.control`：表示使用之前设置好的交叉验证控制对象`cv.control`进行交叉验证。
- `tuneLength = 15`：表示对决策树的参数`cp`进行调优，尝试15个不同的取值。`cp`参数是决策树中的复杂度参数，用于控制树的剪枝程度。

最终，该代码将返回在交叉验证中性能最佳的`rpart`决策树模型，存储在`rpart.cv`对象中。可以通过访问`rpart.cv`对象的属性来获取关于最佳模型的信息，如最佳参数设置、训练误差等。

````r
rpart.cv
````

````r
# Plot model selection
plot(rpart.cv)
````
