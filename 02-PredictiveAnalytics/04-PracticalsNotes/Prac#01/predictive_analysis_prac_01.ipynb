```r
library(tidyverse)  # 导入必备的程序包
```

```r
census.data <- read.csv("camden_census_data.csv")  # 读入数据
```

```r
# 查看一下这个数据的信息
head(census.data)  # 前6个
```

```r
# head()函数
head(census.data, n = 10)  # 指定前10个
```

```r
# ncol()函数
ncol(census.data)  # 计算有多少列
```

```r
# nrow()函数
nrow(census.data)  # 计算有多少行
```

```r
# names()函数
names(census.data)  # 计算有多少variables/features
```

```r
# 修改变OA这个名称
names(census.data)[1] <- "Output_Area"
names(census.data)
```

```r
# calculate mean for unemployed
mean(census.data$Unemployed)
```

```r
# calculate median for unemployed
median(census.data$Unemployed)
```

```r
# calculate standard deviation for Unemployed
sd(census.data$Unemployed)
```

```r
# calculate range for Unemployed
range(census.data$Unemployed)
```

```r
# provide a statistic descriptive
summary(census.data)
```

```r
# create a histogram
hist(census.data$Unemployed)
```

```r
hist(census.data$Unemployed, breaks = 20, col = "blue", main = "% in full-time")
```

```r
# create a boxplot diagram
boxplot(census.data$Unemployed, breaks = 20, col = "red", main = "% in full-time")
```

```r
# box and whisker plots
boxplot(census.data[, 2, 5], xlab = "Percentage")
```
