**PyMuPDF + pyttesseract**

```python
import fitz  # PyMuPDF
from PIL import Image
import pytesseract
import io

#Specify the installation path of Tesseract and modify it according to your actual installation location
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Windows
# pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'  # Linux

def pdf_to_text(pdf_path):
    #Open PDF file
    doc = fitz.open(pdf_path)
    text = ''
    
    for page_num in range(len(doc)):
        # Get the image list of a single page
        page = doc.load_page(page_num)
        pix = page.get_pixmap()
        img = Image.open(io.BytesIO(pix.tobytes()))

        # Use OCR to extract text from images
        page_text = pytesseract.image_to_string(img, lang='eng')
        text += page_text + '\n'
    
    return text

pdf_path = '236-020-SAS-120_0.pdf'  # Please modify the path according to the actual situation
extracted_text = pdf_to_text(pdf_path)
print(extracted_text)
```

**PyMuPDF + Keras_OCR**

```python
import fitz #PyMuPDF
import keras_ocr

# Initialize the pipeline of keras-ocr
pipeline = keras_ocr.pipeline.Pipeline()

#Open PDF file
pdf_path = '236-020-SAS-120_0.pdf'
pdf_document = fitz.open(pdf_path)

# Traverse each page of the PDF document
for page_num in range(len(pdf_document)):
     # Get the page
     page = pdf_document.load_page(page_num)
    
     # Convert PDF pages to images (pix)
     pix = page.get_pixmap()
    
     # Save image directly to file
     image_path = f"page_{page_num + 1}.png"
     pix.save(image_path)
    
     # Use keras-ocr to read the saved image and recognize the text
     image = keras_ocr.tools.read(image_path)
     predictions = pipeline.recognize([image])[0]
    
     #Print prediction results
     print(f"Page {page_num + 1}:")
     for text, _ in predictions:
         print(text)

pdf_document.close()
```

```python
import fitz #PyMuPDF
import easyocr
import numpy as np
from PIL import Image
from io import BytesIO  # Add this line

# Create a reader instance of easy-ocr
reader = easyocr.Reader(['en']) # You can add other languages as needed

# Open PDF file
pdf_path = '236-020-SAS-120_0.pdf'
pdf_document = fitz.open(pdf_path)

# Traverse each page of the PDF document
for page_num in range(len(pdf_document)):
    # Get the page
    page = pdf_document.load_page(page_num)

    # Convert PDF pages to images
    pix = page.get_pixmap()

    # Convert the image to byte array and load it into the PIL Image object
    img = Image.open(BytesIO(pix.tobytes()))

    # Convert PIL Image to numpy array, because easy-ocr requires images in numpy array format
    np_img = np.array(img)

    # Use easy-ocr to identify text in images
    results = reader.readtext(np_img)

    # Print recognition results
    print(f"Page {page_num + 1}:")
    for (bbox, text, prob) in results:
        print(f"Detected text: {text} with confidence {prob}")

pdf_document.close()
```
