**PyMuPDF + pyttesseract**

```python
import fitz  # PyMuPDF
from PIL import Image
import pytesseract
import io

#Specify the installation path of Tesseract and modify it according to your actual installation location
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Windows
# pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'  # Linux

def pdf_to_text(pdf_path):
    #Open PDF file
    doc = fitz.open(pdf_path)
    text = ''
    
    for page_num in range(len(doc)):
        # Get the image list of a single page
        page = doc.load_page(page_num)
        pix = page.get_pixmap()
        img = Image.open(io.BytesIO(pix.tobytes()))

        # Use OCR to extract text from images
        page_text = pytesseract.image_to_string(img, lang='eng', config='--psm 6')
        text += page_text + '\n'
    
    return text

pdf_path = r'pdf_files\236-020-STR-001_D.pdf'  # Please modify the path according to the actual situation
extracted_text = pdf_to_text(pdf_path)
print(extracted_text)
```

**PyMuPDF + Keras_OCR**

```python
import fitz #PyMuPDF
import keras_ocr

# Initialize the pipeline of keras-ocr
pipeline = keras_ocr.pipeline.Pipeline()

#Open PDF file
pdf_path = '236-020-SAS-120_0.pdf'
pdf_document = fitz.open(pdf_path)

# Traverse each page of the PDF document
for page_num in range(len(pdf_document)):
     # Get the page
     page = pdf_document.load_page(page_num)
    
     # Convert PDF pages to images (pix)
     pix = page.get_pixmap()
    
     # Save image directly to file
     image_path = f"page_{page_num + 1}.png"
     pix.save(image_path)
    
     # Use keras-ocr to read the saved image and recognize the text
     image = keras_ocr.tools.read(image_path)
     predictions = pipeline.recognize([image])[0]
    
     #Print prediction results
     print(f"Page {page_num + 1}:")
     for text, _ in predictions:
         print(text)

pdf_document.close()
```

```python
import fitz #PyMuPDF
import easyocr
import numpy as np
from PIL import Image
from io import BytesIO  # Add this line

# Create a reader instance of easy-ocr
reader = easyocr.Reader(['en']) # You can add other languages as needed

# Open PDF file
pdf_path = '236-020-SAS-120_0.pdf'
pdf_document = fitz.open(pdf_path)

# Traverse each page of the PDF document
for page_num in range(len(pdf_document)):
    # Get the page
    page = pdf_document.load_page(page_num)

    # Convert PDF pages to images
    pix = page.get_pixmap()

    # Convert the image to byte array and load it into the PIL Image object
    img = Image.open(BytesIO(pix.tobytes()))

    # Convert PIL Image to numpy array, because easy-ocr requires images in numpy array format
    np_img = np.array(img)

    # Use easy-ocr to identify text in images
    results = reader.readtext(np_img)

    # Print recognition results
    print(f"Page {page_num + 1}:")
    for (bbox, text, prob) in results:
        print(f"Detected text: {text} with confidence {prob}")

pdf_document.close()
```

```python
import cv2
import pytesseract
from pytesseract import Output
import numpy as np

# 载入图片
image = cv2.imread('testing.png')

# 转换为灰度图
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 使用阈值化来使图像对比更明显
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

# 寻找轮廓
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

extracted_texts_psm6 = []

for contour in contours:
    # 获取轮廓的边界框
    x, y, w, h = cv2.boundingRect(contour)
    
    # 仅当宽度和高度大于某个阈值时才考虑边界框（过滤掉太小的轮廓）
    if w > 50 and h > 20:
        # 提取边界框内的图像区域
        roi = image[y:y+h, x:x+w]
        
        # 使用 '--psm 6' 参数
        text = pytesseract.image_to_string(roi, lang='eng', config='--psm 6')
        extracted_texts_psm6.append(text.strip())

print(extracted_texts_psm6)
```

```python
import fitz  # PyMuPDF
from PIL import Image
import pytesseract
import io

#Specify the installation path of Tesseract and modify it according to your actual installation location
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Windows
# pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'  # Linux

def extract_text_from_area(pdf_path, coordinates, page_num=0):
    # Open PDF file
    doc = fitz.open(pdf_path)
    
    # Ensure the specified page number is within the document
    if page_num < 0 or page_num >= len(doc):
        return "Page number out of range."
    
    # Load the specific page
    page = doc.load_page(page_num)
    
    # Extracting specific areas from the PDF
    extracted_texts = {}
    for key, (x0, y0, x1, y1) in coordinates.items():
        # Define rectangle for area of interest
        rect = fitz.Rect(x0, y0, x1, y1)
        # Get the clip of the area
        clip = page.get_pixmap(clip=rect)
        # Convert clip to an image
        img = Image.open(io.BytesIO(clip.tobytes()))
        # Use OCR to extract text from the image
        text = pytesseract.image_to_string(img, lang='eng', config='--psm 6')
        extracted_texts[key] = text
    
    return extracted_texts

pdf_path = r'pdf_files\236-020-SAS-120_0.pdf'  # Please modify the path according to the actual situation
coordinates = {
    "title": (62, 62, 780, 174),  # Title and subtitle
    "notes": (62, 1490, 890, 1620),  # Notes
    "information_block": (1488, 1498, 2327, 1627)  # Information block
}
extracted_texts = extract_text_from_area(pdf_path, coordinates)
for key, text in extracted_texts.items():
    print(f"{key}:\n{text}\n")
```

```python
import pytesseract
from PIL import Image
import cv2
import numpy as np

# 加载图片
image_path = 'testing.png'
image = Image.open(image_path)

# 使用OpenCV转换为灰度图像
image_np = np.array(image)
gray = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)

# 通过提高对比度和应用阈值来优化图像
# 可以试验不同的阈值以找到最佳结果
enhanced = cv2.equalizeHist(gray)
_, binary = cv2.threshold(enhanced, 128, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 使用PIL处理优化后的图像
processed_image = Image.fromarray(binary)

# 对处理后的图像应用OCR
custom_config = r'--oem 3 --psm 6'
extracted_text = pytesseract.image_to_string(processed_image, config=custom_config)

extracted_text
```
