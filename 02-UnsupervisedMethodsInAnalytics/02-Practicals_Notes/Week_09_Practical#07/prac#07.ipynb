**Assume that the given data follows a normal distribution. Use the 3-sigma method to detect if there are outliers in this dataset. List the outlier(s) and present the steps in your answer (e.g., how you calculated / what approach you used to detect the outlier(s)). (Hint: calculate the mean and standard deviation of the data set and decide the outlier using the individual value’s z-score)**

To detect outliers using the 3-sigma method, we need to calculate the mean and standard deviation of the dataset and then determine the z-scores of individual values. The z-score measures how many standard deviations an individual value is away from the mean. Generally, values with a z-score greater than 3 or less than -3 are considered outliers.

Let's calculate the outliers using the given dataset "Prac7-data.csv":

Calculate the mean (μ) of the dataset:
μ = (1.7 + 1.77 + 1.68 + ... + 1.59) / 20

Calculate the standard deviation (σ) of the dataset:
σ = sqrt(((1.7 - μ)^2 + (1.77 - μ)^2 + ... + (1.59 - μ)^2) / 20)

Calculate the z-score for each value:
z-score = (value - μ) / σ

Identify outliers:
Any value with a z-score greater than 3 or less than -3 is considered an outlier.

Now let's write the Python code to perform these calculations:

```python
import pandas as pd
import numpy as np

# Read the dataset from CSV
df = pd.read_csv("Prac7-data.csv")

# Calculate the mean and standard deviation
mean = np.mean(df['value'])
std_dev = np.std(df['value'])

# Calculate z-scores for each value
df['z_score'] = (df['value'] - mean) / std_dev

# Find outliers
outliers = df[(df['z_score'] > 3) | (df['z_score'] < -3)]

# Print the outliers
print("Outliers:")
print(outliers)
```
