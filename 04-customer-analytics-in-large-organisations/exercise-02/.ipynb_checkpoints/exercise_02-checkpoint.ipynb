```python
# Convert Sale_Date to datetime format for accurate processing
ice_cream_data['Sale_Date'] = pd.to_datetime(ice_cream_data['Sale_Date'])

# Calculate daily statistics
# Total and average quantity of ice cream sold per day
daily_sales = ice_cream_data.groupby('Sale_Date')['Quantity_Sold'].sum()
average_daily_sales = daily_sales.mean()

# Number of unique customers (receipts) per day
daily_customers = ice_cream_data.groupby('Sale_Date')['UniSA_Receipt_No1'].nunique()
average_daily_customers = daily_customers.mean()

# Average sales per visit (by receipt)
sales_per_visit = ice_cream_data.groupby(['Sale_Date', 'UniSA_Receipt_No1'])['Quantity_Sold'].sum()
average_sales_per_visit = sales_per_visit.mean()

average_daily_sales, average_daily_customers, average_sales_per_visit
```

```python
# Calculate average sales per visit per day
average_sales_per_visit_per_day = sales_per_visit.groupby('Sale_Date').mean()

# Merging the corrected daily statistics into a single DataFrame
combined_daily_stats_corrected = pd.DataFrame({
    'Total_Ice_Cream_Sold': daily_sales,
    'Unique_Customers': daily_customers,
    'Average_Sales_Per_Visit': average_sales_per_visit_per_day
}).reset_index()

combined_daily_stats_corrected.head()
```

```python
from scipy.stats import lognorm
import numpy as np

# Fit a log-normal distribution
# We need to calculate the natural log of sales to fit a log-normal distribution
log_sales = np.log(daily_sales)

# Estimate parameters for the log-normal distribution: mean and standard deviation of the log of the data
shape, loc, scale = lognorm.fit(daily_sales, floc=0)  # floc=0 to force fit loc to 0 for simplicity

# Create a lognormal distribution with estimated parameters
lognormal_dist = lognorm(shape, loc, scale)

# Calculate the 5%, 1%, 95%, and 99% quantiles from the fitted log-normal distribution
fitted_percentiles = lognormal_dist.ppf([0.01, 0.05, 0.95, 0.99])

fitted_percentiles
```

```python
# Set up the plot with empirical and fitted distribution data
plt.figure(figsize=(10, 6))
sns.histplot(daily_sales, bins=30, kde=False, color='skyblue', label='Empirical Data')
sns.lineplot(x=np.linspace(0, 800, 1000), y=lognormal_dist.pdf(np.linspace(0, 800, 1000)) * len(daily_sales) * (800 / 30), color='red', label='Fitted Log-normal Distribution')

# Add percentile lines to the histogram
plt.axvline(x=fitted_percentiles[0], color='purple', linestyle='--', label='1% Fitted (Very low risk)')
plt.axvline(x=fitted_percentiles[1], color='red', linestyle='--', label='5% Fitted (Low risk)')
plt.axvline(x=fitted_percentiles[2], color='green', linestyle='--', label='95% Fitted (High risk)')
plt.axvline(x=fitted_percentiles[3], color='orange', linestyle='--', label='99% Fitted (Very high risk)')

# Adding labels and title
plt.title('Distribution of Daily Ice Cream Sales with Fitted Log-normal Distribution and its Risk Percentiles')
plt.xlabel('Daily Ice Cream Sales')
plt.ylabel('Frequency / Probability Density')
plt.legend()

plt.show()
```

```python
import numpy as np

# Setting the number of days to simulate (10 years * 365 days)
num_days = 3650

# Initialize an empty list to store the simulated total sales per day
simulated_sales = []

# Bootstrapping process
np.random.seed(42)  # Set seed for reproducibility
for _ in range(num_days):
    # Randomly choose a number of customers for the day based on empirical distribution
    num_customers_today = np.random.choice(daily_customers.values)
    
    # Randomly sample sales for these customers based on empirical sales per visit data
    if num_customers_today > 0:
        sales_today = np.random.choice(sales_per_visit.values, size=num_customers_today)
        total_sales_today = np.sum(sales_today)
    else:
        total_sales_today = 0
    
    # Append the total sales for the day to the list`
    simulated_sales.append(total_sales_today)

# Convert the list to a numpy array for further analysis
simulated_sales = np.array(simulated_sales)

# Calculate and print the 1% and 5% quantiles (worst case scenarios)
quantile_1 = np.percentile(simulated_sales, 1)
quantile_5 = np.percentile(simulated_sales, 5)
quantile_95 = np.percentile(simulated_sales, 95)
quantile_99 = np.percentile(simulated_sales, 99)

quantile_1, quantile_5, quantile_95, quantile_99
```

```python
import numpy as np
import pandas as pd
from scipy.stats import poisson, lognorm

# Parameters based on analysis
weekday_mean_customers = 150
weekend_mean_customers = 250
shape, loc, scale = 0.415, 0, 1.352  # Adjust these parameters based on actual data analysis

# Assume we have daily weather data influencing sales
weather_impact = np.random.normal(1.0, 0.1, days)  # Random normal multiplier for weather

# Initialize random seed for reproducibility
np.random.seed(42)

# Simulating sales for 10 years, considering weekends
total_sales = []
days = 365 * 10  # Simulating for 10 years
start_date = pd.Timestamp('2016-01-01')

for day in range(days):
    current_date = start_date + pd.Timedelta(days=day)
    if current_date.dayofweek < 5:  # Weekdays
        num_customers = poisson.rvs(weekday_mean_customers)
    else:  # Weekends
        num_customers = poisson.rvs(weekend_mean_customers)
    
    # Adjust sales volume based on weather
    sales_per_customer = lognorm.rvs(shape, loc, scale * weather_impact[day], size=num_customers)
    total_daily_sales = np.sum(sales_per_customer)
    total_sales.append(total_daily_sales)

# Analysis of results
total_sales = np.array(total_sales)
high_risk_threshold = np.percentile(total_sales, 95)
low_risk_threshold = np.percentile(total_sales, 5)

print(f"High Risk Threshold: {high_risk_threshold}")
print(f"Low Risk Threshold: {low_risk_threshold}")
```
