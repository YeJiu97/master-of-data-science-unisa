# Deal with Dataset

```python
import pandas as pd

# Load the dataset
file_path = './output/df_ice_cream_all.csv'
df_ice_cream = pd.read_csv(file_path)
```

# Question #01

```python
# Step 1: Convert 'Sale_Date' to datetime format
df_ice_cream['Sale_Date'] = pd.to_datetime(df_ice_cream['Sale_Date'])

# Step 2: Handle negative and extreme values in 'Quantity_Sold'
# Removing negative quantities and capping extreme values at a reasonable upper limit (e.g., 99th percentile)
quantile_99 = df_ice_cream['Quantity_Sold'].quantile(0.99)
df_ice_cream = df_ice_cream[(df_ice_cream['Quantity_Sold'] > 0) & (df_ice_cream['Quantity_Sold'] <= quantile_99)]

# Step 3: Calculate daily statistics
# Total quantity of ice cream sold per day
daily_sales = df_ice_cream.groupby('Sale_Date')['Quantity_Sold'].sum()

# Number of customers buying ice cream per day
daily_customers = df_ice_cream.groupby('Sale_Date')['UniSA_Receipt_No1'].nunique()

# Average sales of ice cream per customer per visit (average per day)
avg_sales_per_customer = daily_sales / daily_customers

# Prepare to display the results
daily_statistics = pd.DataFrame({
    'Total Quantity Sold': daily_sales,
    'Number of Customers': daily_customers,
    'Average Sales per Customer': avg_sales_per_customer
})

daily_statistics.describe()  # Show summary statistics to get the average daily metrics
```

```python
import matplotlib.pyplot as plt

# Create histograms for the three variables
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Histogram for Total Quantity Sold
axes[0].hist(daily_statistics['Total Quantity Sold'], bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Distribution of Total Quantity Sold')
axes[0].set_xlabel('Total Quantity Sold')
axes[0].set_ylabel('Frequency')

# Histogram for Number of Customers
axes[1].hist(daily_statistics['Number of Customers'], bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title('Distribution of Number of Customers')
axes[1].set_xlabel('Number of Customers')
axes[1].set_ylabel('Frequency')

# Histogram for Average Sales per Customer
axes[2].hist(daily_statistics['Average Sales per Customer'], bins=30, color='salmon', edgecolor='black')
axes[2].set_title('Distribution of Average Sales per Customer')
axes[2].set_xlabel('Average Sales per Customer')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Empirical Data

```python
# Calculate the 95% and 99% quantiles for the total quantity of ice cream sold per day
quantile_05 = daily_sales.quantile(0.05)
quantile_01 = daily_sales.quantile(0.01)
quantile_95 = daily_sales.quantile(0.95)
quantile_99 = daily_sales.quantile(0.99)

# Visualizing the distribution and all four quantiles
plt.figure(figsize=(10, 6))
plt.hist(daily_sales, bins=30, color='lightblue', edgecolor='black', alpha=0.7)
plt.axvline(quantile_05, color='red', linestyle='dashed', linewidth=1, label=f'5% Quantile = {quantile_05:.2f}')
plt.axvline(quantile_01, color='green', linestyle='dashed', linewidth=1, label=f'1% Quantile = {quantile_01:.2f}')
plt.axvline(quantile_95, color='blue', linestyle='dashed', linewidth=1, label=f'95% Quantile = {quantile_95:.2f}')
plt.axvline(quantile_99, color='purple', linestyle='dashed', linewidth=1, label=f'99% Quantile = {quantile_99:.2f}')
plt.title('Distribution of Total Ice Cream Sales Per Day with Various Risk Quantiles')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

不完全是正态分布，所以这里尝试使用不同的分布来选择出一个最佳的分布：

```python
from scipy import stats
import numpy as np

# Fit different distributions to the data
distributions = {
    "normal": stats.norm,
    "exponential": stats.expon,
    "log-normal": stats.lognorm,
    "gamma": stats.gamma
}

# Correcting the comparison between the fitted PDF and the empirical distribution
fit_results_corrected = {}
hist_bins = np.histogram(daily_sales, bins=30, density=True)
hist_bin_edges = hist_bins[1]
hist_values = hist_bins[0]
bin_centers = (hist_bin_edges[:-1] + hist_bin_edges[1:]) / 2

for name, dist in distributions.items():
    # Fit distribution to the data
    params = dist.fit(daily_sales)
    
    # Separate parts of parameters
    arg = params[:-2]
    loc = params[-2]
    scale = params[-1]
    
    # Calculate fitted PDF at the bin centers
    pdf_values = dist.pdf(bin_centers, *arg, loc=loc, scale=scale)
    
    # Calculate the sum of squared errors (SSE)
    sse = np.sum((hist_values - pdf_values) ** 2)
    
    # Store results
    fit_results_corrected[name] = {
        "params": params,
        "sse": sse
    }

fit_results_corrected
```

这里使用log-normal：

```python
# Extract parameters for the best fitting distribution (Log-Normal)
shape, loc, scale = fit_results_corrected['log-normal']['params']

# Create a log-normal distribution with the fitted parameters
log_normal_dist = stats.lognorm(s=shape, loc=loc, scale=scale)

# Calculate the quantiles for the fitted log-normal distribution
lognorm_quantile_01 = log_normal_dist.ppf(0.01)
lognorm_quantile_05 = log_normal_dist.ppf(0.05)
lognorm_quantile_95 = log_normal_dist.ppf(0.95)
lognorm_quantile_99 = log_normal_dist.ppf(0.99)

# Visualize the fitted distribution and the calculated quantiles
plt.figure(figsize=(10, 6))
x = np.linspace(min(daily_sales), max(daily_sales), 1000)
pdf = log_normal_dist.pdf(x)
plt.plot(x, pdf, 'r-', lw=2, label='Log-normal fit')
plt.hist(daily_sales, bins=30, density=True, alpha=0.6, color='lightblue', edgecolor='black')
plt.axvline(lognorm_quantile_01, color='green', linestyle='dashed', linewidth=1, label=f'1% Quantile = {lognorm_quantile_01:.2f}')
plt.axvline(lognorm_quantile_05, color='red', linestyle='dashed', linewidth=1, label=f'5% Quantile = {lognorm_quantile_05:.2f}')
plt.axvline(lognorm_quantile_95, color='blue', linestyle='dashed', linewidth=1, label=f'95% Quantile = {lognorm_quantile_95:.2f}')
plt.axvline(lognorm_quantile_99, color='purple', linestyle='dashed', linewidth=1, label=f'99% Quantile = {lognorm_quantile_99:.2f}')
plt.title('Fitted Log-Normal Distribution with Risk Quantiles')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Density')
plt.legend()
plt.show()

(lognorm_quantile_01, lognorm_quantile_05, lognorm_quantile_95, lognorm_quantile_99)
```

# Question #03

## Boottrapping

```python
import numpy as np

# Constants
days_per_year = 365
simulation_years = 10
total_days = days_per_year * simulation_years

# Generate the number of customers per day from empirical distribution
np.random.seed(99)  # for reproducibility
simulated_customers_per_day = np.random.choice(daily_customers, size=total_days, replace=True)

# Randomly sample sales amounts per customer from the empirical distribution of sales per item
simulated_daily_sales = []

for day in range(total_days):
    # Number of customers for the day
    num_customers = simulated_customers_per_day[day]
    
    # Sample the sales for each customer
    if num_customers > 0:
        sales_per_customer = np.random.choice(df_ice_cream['Quantity_Sold'], size=num_customers, replace=True)
        total_sales = np.sum(sales_per_customer)
    else:
        total_sales = 0
    
    simulated_daily_sales.append(total_sales)

# Convert to numpy array for further analysis
simulated_daily_sales = np.array(simulated_daily_sales)

# Calculate the 1% and 5% quantiles
simulated_quantile_01 = np.percentile(simulated_daily_sales, 1)
simulated_quantile_05 = np.percentile(simulated_daily_sales, 5)
simulated_quantile_95 = np.percentile(simulated_daily_sales, 95)
simulated_quantile_99 = np.percentile(simulated_daily_sales, 99)

(simulated_quantile_01, simulated_quantile_05, simulated_quantile_95, simulated_quantile_99)
```

```python
# Visualizing the simulated daily sales distribution with quantiles
plt.figure(figsize=(10, 6))
plt.hist(simulated_daily_sales, bins=30, color='lightblue', edgecolor='black', alpha=0.7)
plt.axvline(simulated_quantile_01, color='green', linestyle='dashed', linewidth=1, label=f'1% Quantile = {simulated_quantile_01}')
plt.axvline(simulated_quantile_05, color='red', linestyle='dashed', linewidth=1, label=f'5% Quantile = {simulated_quantile_05}')
plt.axvline(simulated_quantile_95, color='blue', linestyle='dashed', linewidth=1, label=f'95% Quantile = {simulated_quantile_95}')
plt.axvline(simulated_quantile_99, color='purple', linestyle='dashed', linewidth=1, label=f'99% Quantile = {simulated_quantile_99}')
plt.title('Simulated Distribution of Daily Ice Cream Sales (10 years)')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
