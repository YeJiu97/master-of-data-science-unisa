**Take the solar radiation data from File a, and copy it into File c, and run the 
power spectrum tool to find out which frequencies are important**.

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks  # Import the find_peaks function

def calculate_dft(data, num_frequencies):
    num_data = len(data)
    n2 = num_data / 2
    ret = {'frequency': [], 'ai': [], 'bi': [], 'power': []}
    
    for i in range(num_frequencies):
        s1 = 0
        s2 = 0
        s0 = sum(data)
        
        for j in range(1, num_data):
            angle = math.pi * i / n2 * (j + 1)
            cit = math.cos(angle)
            sit = math.sin(angle)
            s1 += cit * data[j]
            s2 += sit * data[j]
        
        s1 /= n2
        s2 /= n2
        
        ret['frequency'].append(i)
        ret['ai'].append(s1)
        ret['bi'].append(s2)
        ret['power'].append(s1 ** 2 + s2 ** 2)
    
    ret['ai'][0] = sum(data) / len(data)
    return pd.DataFrame(ret)
```

```python
solar_radia2017 = pd.read_excel('HalfHourSolarRadiation2017.xlsx')
    
# Convert half-hourly data to daily data
solar_radia2017['group'] = np.floor(solar_radia2017.index / 48)
daily_data = solar_radia2017.groupby('group').sum()
    
num_frequencies = 100
daily_power_spectrum = calculate_dft(daily_data.GHI, num_frequencies)
    
peaks, _ = find_peaks(daily_power_spectrum['power'][1:], height=0)
important_frequencies = daily_power_spectrum['frequency'][peaks + 1]
    
print("Important Frequencies:")
for freq in important_frequencies:
    print(f"Frequency: {freq}, Power: {daily_power_spectrum['power'][freq + 1]}")
    
plt.bar(daily_power_spectrum['frequency'][1:], daily_power_spectrum['power'][1:])
plt.plot(important_frequencies, daily_power_spectrum['power'][peaks + 1], 'ro', label='Important Frequencies')
plt.title('Daily Power Spectrum with Important Frequencies', fontsize=14)
plt.xlabel('Frequency', fontsize=14)
plt.ylabel('Amplitude', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```

**Use File c to find the Fourier series model for the seasonality. Note that the 
Template is designed for hourly data. You will have to make some adjustment 
to use it for half hourly data plus change the relevant frequencies if necessar**y.

```python
import numpy as np

# Define time intervals for half-hourly data
num_intervals = len(daily_data)
time_intervals = np.arange(num_intervals) * 0.5  # Each interval is half-hourly

# Define Fourier series components based on important frequencies
def fourier_series_components(t, frequencies):
    components = []
    for freq in frequencies:
        w = 2 * np.pi * freq / (num_intervals * 0.5)  # Adjust frequency for half-hourly data
        component = np.cos(w * t) + np.sin(w * t)
        components.append(component)
    return components

# Calculate Fourier series model for seasonality
def calculate_fourier_series(data, frequencies):
    model = np.zeros_like(data)
    components = fourier_series_components(time_intervals, frequencies)
    
    for i in range(len(frequencies)):
        model += components[i] * data.max() * (daily_power_spectrum['ai'][frequencies[i]] ** 2 + daily_power_spectrum['bi'][frequencies[i]] ** 2)
    
    return model

# Define important frequencies (use the ones you obtained)
important_frequencies = [4, 8, 12, 16, 19, 22, 26, 29, 35, 39, 41, 43, 45, 48, 51, 53, 55, 57, 61, 64, 66, 69, 72, 75, 82, 85, 87, 89, 91, 93, 96]

# Calculate Fourier series model for seasonality
seasonality_model = calculate_fourier_series(daily_data['GHI'], important_frequencies)

# Plot original data and seasonality model
plt.plot(time_intervals, daily_data['GHI'], label='Original Data')
plt.plot(time_intervals, seasonality_model, label='Seasonality Model')
plt.title('Original Data vs. Seasonality Model', fontsize=14)
plt.xlabel('Time (hours)', fontsize=14)
plt.ylabel('GHI', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

**Take the difference between the data and the Fourier model - the residuals - 
and take them to Minitab and find the best ARMA(p,q) model**.y.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pmdarima.arima import auto_arima

# Load your data and compute the Fourier model (as in the previous code)

# Calculate the residuals
residuals = daily_data['GHI'] - seasonality_model

# Plot the residuals
plt.plot(time_intervals, residuals)
plt.title('Residuals', fontsize=14)
plt.xlabel('Time (hours)', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.grid(True)
plt.show()

# Check if residuals are stationary using ADF test
adf_result = adfuller(residuals)
print("ADF Test Results:")
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])
print("Critical Values:", adf_result[4])

# Plot ACF and PACF of residuals
plot_acf(residuals, lags=20)
plt.title('ACF of Residuals', fontsize=14)
plt.xlabel('Lags', fontsize=14)
plt.ylabel('Autocorrelation', fontsize=14)
plt.grid(True)
plt.show()

plot_pacf(residuals, lags=20)
plt.title('PACF of Residuals', fontsize=14)
plt.xlabel('Lags', fontsize=14)
plt.ylabel('Partial Autocorrelation', fontsize=14)
plt.grid(True)
plt.show()

# Find the best ARMA(p,q) model using pmdarima
model = auto_arima(residuals, seasonal=False, suppress_warnings=True)
print("Best ARMA Model (p, d, q):", model.order)

# Fit the best ARMA model
best_arma_model = model.fit(residuals)

# Print ARMA model summary
print(best_arma_model.summary())
```

**Use the ARMA model to forecast one step ahead for the residuals and add that 
to the Fourier series model to get the full one step ahead forecast**.and q.

```python
# Forecast one step ahead using the ARMA model
arma_forecast = best_arma_model.get_forecast(steps=1)
arma_forecast_mean = arma_forecast.predicted_mean.values[0]

# Add the ARMA forecast to the Fourier series model
one_step_ahead_forecast = seasonality_model[-1] + arma_forecast_mean

# Plot the original data, the Fourier model, and the one step ahead forecast
plt.plot(time_intervals, daily_data['GHI'], label='Original Data')
plt.plot(time_intervals, seasonality_model, label='Seasonality Model')
plt.plot(time_intervals[-1] + 0.5, one_step_ahead_forecast, 'ro', label='One Step Ahead Forecast')
plt.title('Original Data, Seasonality Model, and One Step Ahead Forecast', fontsize=14)
plt.xlabel('Time (hours)', fontsize=14)
plt.ylabel('GHI', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Calculate the errors between the actual data and the one-step ahead forecast
actual_data = daily_data['GHI'].values
forecast_error = actual_data[-1] - one_step_ahead_forecast

# Calculate error metrics
mae = mean_absolute_error(actual_data[-1], one_step_ahead_forecast)
mse = mean_squared_error(actual_data[-1], one_step_ahead_forecast)
rmse = np.sqrt(mse)
mape = np.abs(forecast_error / actual_data[-1]) * 100

# Print error metrics
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Percentage Error (MAPE):", mape)

# You can also plot the actual data and the one-step ahead forecast
plt.plot(time_intervals, actual_data, label='Actual Data')
plt.plot(time_intervals[-1] + 0.5, one_step_ahead_forecast, 'ro', label='One Step Ahead Forecast')
plt.title('Actual Data vs. One Step Ahead Forecast', fontsize=14)
plt.xlabel('Time (hours)', fontsize=14)
plt.ylabel('GHI', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
